```python
# load all three analogy files
# from '../data/hmn_analogies.txt'
# '../data/llm_analogies.txt'
# '../data/vlm_analogies.txt'
import csv
import random
random.seed(1337)

save_path = '../data/random_analogy_list_testSet'

paths = ['../data/hmn_analogies.txt', '../data/llm_analogies.txt', '../data/vlm_analogies.txt']

hmn_analogies = []
llm_analogies = []
vlm_analogies = []


for path in paths:
    with open(path, 'r') as f:
        for line in f:
            if path == paths[0]:
                hmn_analogies.append(line.strip())
            elif path == paths[1]:
                llm_analogies.append(line.strip())
            elif path == paths[2]:
                vlm_analogies.append(line.strip())
```

```python
def split_preserve_sentences(line):
    return next(csv.reader([line], skipinitialspace=True))

hmn_words = {'words_1to10': [], 'words_11to20': [], 'words_21to30': []}

for i, line in enumerate(hmn_analogies):
    line = split_preserve_sentences(line)
    if i < 5:
        for i in range(6, 26, 2):
            hmn_words['words_1to10'].append(line[i])
    elif i < 10:
        for i in range(6, 26, 2):
            hmn_words['words_11to20'].append(line[i])
    else:
        for i in range(6, 26, 2):
            hmn_words['words_21to30'].append(line[i])

```

```python
hmn_sample_old = []
hmn_sample_new = []
hmn_sample_new_new = []
# randmly sample 10 words from each set without duplicates
for key, value in hmn_words.items():
    hmn_sample_old.extend(random.sample(value, 10))
for i, elem in enumerate(hmn_sample_old):
    print(elem)

print("\n\n\n")
# randomly sample 10 words from each set without duplicates
for key, value in hmn_words.items():
    # make sure the new sample does not contain any of the words from the old sample
    hmn_sample_new.extend(random.sample(list(set(value) - set(hmn_sample_old)), 5))
for i, elem in enumerate(hmn_sample_new):
    print(elem)
    
print("\n\n\n")
# randomly sample 10 words from each set without duplicates
for key, value in hmn_words.items():
    # make sure the new sample does not contain any of the words from the old sample
    hmn_sample_new_new.extend(random.sample(list(set(value) - set(hmn_sample_old) - set(hmn_sample_new)), 10))
for i, elem in enumerate(hmn_sample_new_new):
    print(elem)
```

```python
select_from = []
llm_entries = []
for idx, line in enumerate(llm_analogies):
    if idx%4 ==0:
        print(idx, line)
        select_from.append(line.split("analogy:")[1])

    if len(select_from) == 24:
        #print("full")
        llm_entries.append(random.choice(select_from))
        select_from = []

print(len(llm_entries))
```

```python
# load vlm_analogies.txt as json file
import json
vlm_analogies = []
with open('../data/vlm_analogies.txt', 'r') as f:
    for line in f:
        vlm_analogies.append(json.loads(line))


```

```python
vlms_selections = []
for elem in vlm_analogies:
    for key, value in elem.items():
        select_from = []
        for k, v in value.items():
            select_from.append(v[0].split("\n")[0])
        vlms_selections.append(random.choice(select_from))

for x in vlms_selections:
    print(x)

print(len(vlms_selections))
```

```python
# load full_list.txt from the data file

full_list = []
with open('../data/full_list.txt', 'r') as f:
    for i,line in enumerate(f):
        if i < 60:
            full_list.append("'"+line.strip())
        else:
            full_list.append(line.strip())

# shuffle the full_list and print the items
random.shuffle(full_list)

with open("../data/"+save_path+".txt", "w") as f:
    for x in full_list:
        f.write(x+"\n")
```
